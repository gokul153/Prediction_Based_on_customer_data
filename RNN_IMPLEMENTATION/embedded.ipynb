```python
from tensorflow.keras.preprocessing.text  import one_hot
```

```python
# Define the sentences
sentences = [
    "The cat sat on the mat.",
    "I love to play soccer.",
    "The weather is sunny today.",
    "She enjoys reading books.",
    "He is learning to code in Python.",
    "The dog barked at the mailman.",
    "They went hiking in the mountains.",
    "Cooking is my favorite hobby.",
    "Music brings joy to my life.",
    "Traveling opens up new experiences.",
    "football is easy to learn",
    "football is easeist to play"
]
```

```python
sentences
voc_size=10000

```

```python
## define the vocalbary size
one_hot_repr =[one_hot(words,voc_size) for words in sentences]
```

```python
one_hot_repr
```

```python
## word Embedding Representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy
```

```python
sent_length = 8
embedded_docs =pad_sequences(one_hot_repr,padding= 'pre',maxlen=sent_length)
```

```python
embedded_docs
```

```python
## feature represntation 
## i need 10 feautures 
dim =10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length =sent_length))
```

```python
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
