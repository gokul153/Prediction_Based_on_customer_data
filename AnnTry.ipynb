```python
import pandas  as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
data = pd.read_csv('Churn_Modelling.csv')
```

```python
data
## extimated salary is a output statement
```

```python
data = data.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
data.head()
```

```python
label_encoder_gender = LabelEncoder()
data['Gender'] = label_encoder_gender.fit_transform(data['Gender'])
```

```python
data.head()
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo = OneHotEncoder()
```

```python
geo_encoded = onehot_encoder_geo.fit_transform(data[['Geography']]).toarray()
geo_encoded
```

```python
geo_encoded_df = pd.DataFrame(geo_encoded,columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df.head()
```

```python
data
```

```python
## combine one hot encoded column with orginal data 
data = pd.concat([data.drop("Geography",axis=1),geo_encoded_df],axis=1)    
```

```python
data.head()
```

```python
## split data into features and target
x = data.drop('EstimatedSalary',axis=1)
y = data['EstimatedSalary']
```

```python
## split the data in training and testing data
X_train ,X_test,y_train ,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
## scale these feature
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
## save the encoders and scale
with open('label_encoder_gender_sal.pk1','wb') as file:
    pickle.dump(label_encoder_gender,file)
with open('onehot_encoder_geo_sal.pk1','wb') as file:
    pickle.dump(onehot_encoder_geo,file)
```

```python
with open('scaler_sal.pk1','wb') as file:
    pickle.dump(scaler,file)
```

```python
## ANN with regression problem statement 
import tensorflow as tf
from tensorflow.keras.models import  Sequential
from tensorflow.keras.layers import  Dense
```

```python
## build ANN models
model  = Sequential([
    ### Input(shape=(X_train.shape[1],)),  # Define input shape using Input()  chatgpt

    Dense(64,activation='relu',input_shape= (X_train.shape[1],)),  ## hl1
    Dense(32,activation='relu'), ## hiddens layer 2
    Dense(1)   ##Dense  ==1 means linear activation funstion that basically means we are using regression
]

)
```

```python
model.summary()
```

```python
## compile the model
model.compile(optimizer='adam',loss='mean_absolute_error',metrics = ['mae'])
```

```python
model.summary()
```

```python
import datetime
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
log_dir = "regressionlogs/fit" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S") 
tensorflow_callback = TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
## set up Early Stopping
import numpy as np
early_stopping_callback = EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
X_train_array = np.array(X_train)
y_train_array = np.array(y_train)
X_test_array = np.array(X_test)
y_test_array = np.array(y_test)
print(X_train)
```

```python
X_train_array
```

```python
## training model
history =model.fit(
    X_train,y_train,validation_data=(X_test,y_test),
    epochs=100,
    callbacks=[tensorflow_callback,early_stopping_callback]
)
```
